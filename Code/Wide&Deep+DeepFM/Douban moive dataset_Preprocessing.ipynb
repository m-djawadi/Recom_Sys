```python
import pandas as pd
import numpy as np
# 读取电影CSV数据
movies_data = pd.read_csv('./dataset/doubanmovies/movies.csv',  header=0)
movies_data.head()
```

```python
#drop some useless columns
# MOVIE_ID: 电影ID，对应豆瓣的DOUBAN_ID
# NAME: 电影名称
# ALIAS: 别名
# ACTORS: 主演
# COVER: 封面图片地址
# DIRECTORS: 导演
# GENRES: 类型
# OFFICIAL_SITE: 地址
# REGIONS: 制片国家/地区
# LANGUAGES: 语言
# RELEASE_DATE: 上映日期
# MINS: 片长
# IMDB_ID: IMDbID
# DOUBAN_SCORE: 豆瓣评分
# DOUBAN_VOTES: 豆瓣投票数
# TAGS: 标签
# STORYLINE: 电影描述
# SLUG: 加密的url，可忽略
# YEAR: 年份
# ACTOR_IDS: 演员与PERSON_ID的对应关系,多个演员采用“\|”符号分割，格式“演员A:ID\|演员B:ID”；
# DIRECTOR_IDS: 导演与PERSON_ID的对应关系,多个导演采用“\|”符号分割，格式“导演A:ID\|导演B:ID”；
movies_data.drop(columns=['ALIAS', 'COVER','OFFICIAL_SITE','RELEASE_DATE','MINS','IMDB_ID','SLUG','YEAR','ACTOR_IDS','DIRECTOR_IDS'],inplace = True)
movies_data.head()
```

```python
movies_data.dropna(inplace = True)
movies_data.head()
```

```python
# 把电影名做分词，这样学习出来的模型，更有利于提高用户和电影名的匹配度，更容易分析出用户更喜欢那些电影。
# 如用户经常看电影命中有“死” “活”，这样电影“007之你死我活”就更容易被推荐到
import re
import thulac  # this package is to split Chinese vocabulary by vocabulary
thu1 = thulac.thulac()  #默认模式

def SplitChineseSentenceToWords(x):
    wordWithSign = thu1.cut(x, text=True)
    r1 = re.sub(" ._[w]", "", wordWithSign)
    r2 = re.sub("_[\w]+", "", r1)    
    wordWithDelimiter = "|".join(r2.split())   
    return wordWithDelimiter
```

```python
movies_data.loc[:, "MovieNameWord"] = movies_data.loc[:, "NAME"].apply(SplitChineseSentenceToWords)
movies_data.loc[:, "STORYLINEWORD"] = movies_data.loc[:, "STORYLINE"].apply(SplitChineseSentenceToWords)
movies_data.head()
```

```python
# Transfer '/'  to '|' Delimiter
movies_data.loc[:, "ACTORS"] = movies_data.loc[:, "ACTORS"].apply(lambda x : x.replace("/","|"))
movies_data.loc[:, "DIRECTORS"] = movies_data.loc[:, "DIRECTORS"].apply(lambda x : x.replace("/","|"))
movies_data.loc[:, "GENRES"] = movies_data.loc[:, "GENRES"].apply(lambda x : x.replace("/","|"))
movies_data.loc[:, "LANGUAGES"] = movies_data.loc[:, "LANGUAGES"].apply(lambda x : x.replace(" / ","|"))
movies_data.loc[:, "TAGS"] = movies_data.loc[:, "TAGS"].apply(lambda x : x.replace("/","|"))
movies_data.loc[:, "REGIONS"] = movies_data.loc[:, "REGIONS"].apply(lambda x : x.replace(" / ","|"))
```
