```python
#data/Movies_train.csv
samples_data=pd.read_csv('/content/gdrive/My Drive/dl_group/Movies_train.csv')
samples_data=samples_data.drop_duplicates("USER_MD5")
# print(df_movie.head())
samples_data=samples_data.drop("STORYLINEWORD",axis=1)
samples_data=samples_data.drop("STORYLINE",axis=1)
samples_data=samples_data.drop("MovieNameWord",axis=1)
df_movies=samples_data.drop('ACTORS',axis=1)
df_movies=df_movies[:1000]
#df_movies=df_movies[:1000]

"""
Index(['USER_MD5', 'MOVIE_ID', 'RATING', 'NAME', 'ACTORS', 'DIRECTORS',
       'DOUBAN_SCORE', 'DOUBAN_VOTES', 'GENRES', 'LANGUAGES', 'REGIONS',
       'STORYLINE', 'TAGS', 'MovieNameWord', 'STORYLINEWORD'],
      dtype='object')
"""

def expand(df, col, sep=','):
    r = df[col].str.split(sep)
    d = {c: df[c].values.repeat(r.str.len(), axis=0) for c in df.columns}
    d[col] = [i for sub in r for i in sub]
    return pd.DataFrame(d)
df_movies=expand(df_movies, 'GENRES', '|')
df_movies=expand(df_movies, 'TAGS', '|')
df_movies=expand(df_movies, 'LANGUAGES', '|')
df_movies=expand(df_movies, 'REGIONS', '|')
df_movies=expand(df_movies, 'DIRECTORS', '|')

#'DIRECTORS', 'DOUBAN_SCORE', 'DOUBAN_VOTES', 'GENRES', 'LANGUAGES', 'MOVIE_ID', 'NAME', 'RATING', 'REGIONS', 'TAGS', 'USER_MD5'
# print(df_movies.shape) #(6854638, 11)


#-------------------------------------------------------
MOVIE_ID_set=set(df_movies['MOVIE_ID'].unique())
MOVIE_ID_dict={}
for index,cate in enumerate(MOVIE_ID_set):
    MOVIE_ID_dict[cate]=index+1

def MOVIE_ID_2int(MOVIE_ID,MOVIE_ID_dict):
    if MOVIE_ID in MOVIE_ID_dict:
        return MOVIE_ID_dict[MOVIE_ID]
    else:
        return random.choice(list(MOVIE_ID_dict.values()))

#-------------------------------------------------------
USER_MD5_set=set(df_movies['USER_MD5'].unique())
USER_MD5_dict={}
for index,cate in enumerate(USER_MD5_set):
    USER_MD5_dict[cate]=index

def USER_MD5_2int(USER_MD5,USER_MD5_dict):
    if USER_MD5 in USER_MD5_dict:
        return USER_MD5_dict[USER_MD5]
    else:
        return random.choice(list(USER_MD5_dict.values()))
#------------------------------------------------------------

NAME_set=set(df_movies['NAME'].unique())
NAME_dict={}
for index,cate in enumerate(NAME_set):
    NAME_dict[cate]=index

def NAME_2int(NAME,NAME_dict):
    if NAME in NAME_dict:
        return NAME_dict[NAME]
    else:
        return random.choice(list(NAME_dict.values()))
#------------------------------------------------------------

DIRECTORS_set=set(df_movies['DIRECTORS'].unique())
DIRECTORS_dict={}
for index,cate in enumerate(DIRECTORS_set):
    DIRECTORS_dict[cate]=index

def DIRECTORS_2int(DIRECTORS,DIRECTORS_dict):
    if DIRECTORS in DIRECTORS_dict:
        return DIRECTORS_dict[DIRECTORS]
    else:
        return random.choice(list(DIRECTORS_dict.values()))
#------------------------------------------------------------


GENRES_set=set(df_movies['GENRES'].unique())
GENRES_dict={}
for index,cate in enumerate(GENRES_set):
    GENRES_dict[cate]=index

def GENRES_2int(GENRES,GENRES_dict):
    if GENRES in GENRES_dict:
        return GENRES_dict[GENRES]
    else:
        return random.choice(list(GENRES_dict.values()))
#------------------------------------------------------------

LANGUAGES_set=set(df_movies['LANGUAGES'].unique())
LANGUAGES_dict={}
for index,cate in enumerate(LANGUAGES_set):
    LANGUAGES_dict[cate]=index

def LANGUAGES_2int(LANGUAGES,LANGUAGES_dict):
    if LANGUAGES in LANGUAGES_dict:
        return LANGUAGES_dict[LANGUAGES]
    else:
        return random.choice(list(LANGUAGES_dict.values()))
#------------------------------------------------------------

REGIONS_set=set(df_movies['REGIONS'].unique())
REGIONS_dict={}
for index,cate in enumerate(REGIONS_set):
    REGIONS_dict[cate]=index
def REGIONS_2int(REGIONS,REGIONS_dict):
    if REGIONS in REGIONS_dict:
        return REGIONS_dict[REGIONS]
    else:
        return random.choice(list(REGIONS_dict.values()))
#------------------------------------------------------------

TAGS_set=set(df_movies['TAGS'].unique())
TAGS_dict={}
for index,cate in enumerate(TAGS_set):
    TAGS_dict[cate]=index
def TAGS_2int(TAGS,TAGS_dict):
    if TAGS in TAGS_dict:
        return TAGS_dict[TAGS]
    else:
        return random.choice(list(TAGS_dict.values()))
#------------------------------------------------------------
#'DIRECTORS', 'DOUBAN_SCORE', 'DOUBAN_VOTES', 'GENRES', 'LANGUAGES', 'MOVIE_ID', 'NAME', 'RATING', 'REGIONS', 'TAGS', 'USER_MD5'
df_movies['USER_MD5']=df_movies['USER_MD5'].map(lambda x:USER_MD5_2int(x,USER_MD5_dict))
df_movies['MOVIE_ID']=df_movies['MOVIE_ID'].map(lambda x:MOVIE_ID_2int(x,MOVIE_ID_dict))
df_movies['TAGS']=df_movies['TAGS'].map(lambda x:TAGS_2int(x,TAGS_dict))
df_movies['REGIONS']=df_movies['REGIONS'].map(lambda x:REGIONS_2int(x,REGIONS_dict))
df_movies['NAME']=df_movies['NAME'].map(lambda x:NAME_2int(x,NAME_dict))
df_movies['LANGUAGES']=df_movies['LANGUAGES'].map(lambda x:LANGUAGES_2int(x,LANGUAGES_dict))
df_movies['GENRES']=df_movies['GENRES'].map(lambda x:GENRES_2int(x,GENRES_dict))
df_movies['DIRECTORS']=df_movies['DIRECTORS'].map(lambda x:DIRECTORS_2int(x,DIRECTORS_dict))
all_profile=df_movies.to_dict(orient="records")
movie_profile=df_movies[['DIRECTORS', 'GENRES', 'LANGUAGES','MOVIE_ID', 'NAME', 'REGIONS', 'TAGS','RATING']].to_dict(orient="records")

```
