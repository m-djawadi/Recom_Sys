```python
def remove_place_signal(x):
    if '+' in x:
        result=x.split('+')
        if "." in result[0]:
            result1 = result[0].split(".")[0]
            return result1
        else:
            return result[0]
    else:
        if "." in x:
            x=x.split(".")[0]
            return x
        else:
            return x

def convert(x):
    #datetime.strptime(row[0], '%Y-%m-%dT%H:%M:%S').strftime("%Y-%m-%d")
    # d = datetime.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S.%f')
    d = datetime.datetime.strptime(x,"%Y-%m-%d %H:%M:%S")
    t = d.timetuple()
    timeStamp = int(time.mktime(t))
    return timeStamp


df_modcloth['timestamp1']=df_modcloth.timestamp1.apply(lambda x: remove_place_signal(x))
df_modcloth['timestamp'] = df_modcloth.timestamp1.apply(lambda x: convert(x))
df_modcloth=df_modcloth.sort_values(by="timestamp", ascending=True)
```

```python
##'item_id', 'user_id', 'rating', 'timestamp', 'size', 'user_attr','model_attr', 'category', 'brand'
df_modcloth=df_modcloth[:10000]
item_id_set=set(df_modcloth['item_id'].unique())
item_id_dict={}
for index,cate in enumerate(item_id_set):
    item_id_dict[cate]=index

brand_set=set(df_modcloth['brand'].unique())
brand_dict={}
for index,cate in enumerate(brand_set):
    brand_dict[cate]=index

category_set=set(df_modcloth['category'].unique())
category_dict={}
for index,cate in enumerate(category_set):
    category_dict[cate]=index

user_id_set=set(df_modcloth['user_id'].unique())
user_id_dict={}
for index,cate in enumerate(user_id_set):
    user_id_dict[cate]=index

user_attr_set=set(df_modcloth['user_attr'].unique())
user_attr_dict={}
for index,cate in enumerate(user_attr_set):
    user_attr_dict[cate]=index

model_attr_set=set(df_modcloth['model_attr'].unique())
model_attr_dict={}
for index,gender in enumerate(model_attr_set):
    model_attr_dict[gender]=index

def user_attr_2int(user_attr, user_attr_dict):
    if user_attr in user_attr_dict:
        return user_attr_dict[user_attr]
    else:
        return random.choice(list(user_attr_dict.values()))

def item_id_2int(item_id, item_id_dict):
    if item_id in item_id_dict:
        return item_id_dict[item_id]
    else:
        return random.choice(list(item_id_dict.values()))


def model_attr2int(model_attr, model_attr_dict):
    if model_attr in model_attr_dict:
        return model_attr_dict[model_attr]
    else:
        return random.choice(list(model_attr_dict.values()))


def brand_2int(brand_2, brand_dict):
    if brand_2 in brand_dict:
        return brand_dict[brand_2]
    else:
        return random.choice(list(brand_dict.values()))

def user_id_2int(user_id, user_attr_dict):
    if user_id in user_attr_dict:
        return user_attr_dict[user_id]
    else:
        return random.choice(list(user_attr_dict.values()))

def category_2int(category,category_dict):
    if category in category_dict:
        return category_dict[category]
    else:
        return random.choice(list(category_dict.values()))
        
```

```python
#mapping
df_modcloth['model_attr']=df_modcloth['model_attr'].map(lambda x:model_attr2int(x,model_attr_dict))
df_modcloth['brand']=df_modcloth['brand'].map(lambda x:brand_2int(x,brand_dict))
df_modcloth['user_attr']=df_modcloth['user_attr'].map(lambda x:user_attr_2int(x,user_attr_dict))
df_modcloth['user_id']=df_modcloth['user_id'].map(lambda x:user_id_2int(x,user_id_dict))
df_modcloth['category']=df_modcloth['category'].map(lambda x:category_2int(x,category_dict))
df_modcloth['item_id']=df_modcloth['item_id'].map(lambda x:item_id_2int(x,item_id_dict))
df_modcloth=df_modcloth.drop("timestamp1",axis=1)

itemprofile=df_modcloth[['category','brand','item_id']].to_dict(orient='records')
modcloth_profile=df_modcloth.to_dict(orient='records')
```

```python
groupdata=df_modcloth.groupby("user_id")
user_behavior=[]
max=0
for groupname,data in groupdata:
    userkey=groupname
    data=data.to_dict(orient="records")
    item_id={item['item_id'] for item  in data}
    items=list(item_id)
    length = len(items)

    allitems=set(item_id_dict.values())
    seleted1 = set(random.sample(allitems, 10))
    otheritems=allitems.difference(seleted1)
    if length>max:
        max=length
    getnum=10
    seleted2=[]
    seleted2.extend(items)
    samplenum=getnum-length
    if samplenum>0:
        seleted = set(random.sample(otheritems, samplenum))
        seleted2.extend(list(seleted))

    behavior = []

    for item in seleted1:
        if len(behavior) <10:
            userdict = {}
            behavior.append(item)
            hist_len = len(behavior)
            behavior1 = []
            behavior1.extend(behavior)
            if len(seleted2)<1:
                continue
            else:
                itemid = random.choice(seleted2)
                seleted2.remove(itemid)

                for cuitem in modcloth_profile:
                    #'item_id', 'user_id', 'rating', 'timestamp', 'size', 'user_attr','model_attr', 'category', 'brand'
                    if cuitem["user_id"]==groupname:
                        
                        userdict['size']=cuitem['size']
                        userdict['model_attr']=cuitem['model_attr']
                        userdict['user_attr']=cuitem['user_attr']
                        userdict['user_id'] = groupname
                        if cuitem['item_id']==itemid:
                            userdict['brand']=cuitem['brand']
                            userdict['category']=cuitem['category']
                            userdict['rating'] = cuitem['rating']
                            userdict['movie_id'] = itemid
                        else:
                            for item in itemprofile:
                                if item["item_id"] == itemid:
                                    userdict['brand'] = item['brand']
                                    userdict['category'] = item['category']
                                    userdict['rating'] = 0
                                    userdict['movie_id'] = itemid

            if len(behavior1) < 10:
               num = 10 - len(behavior1)
               for i in range(num+1):
                   behavior1.append(0)
               userdict["hist_movie_id"]= " ".join((str(id) for id in behavior1))
               userdict['hist_len']=len(behavior)
               user_behavior.append(userdict)
               continue

        else:
            userdict = {}
            hist_len = 10
            if len(seleted2) < 1:
                continue
            else:
             behavior1 = []
             behavior1.extend(behavior)
             itemid = random.choice(seleted2)
             seleted2.remove(itemid)
             for cuitem in modcloth_profile:
                 # 'item_id', 'user_id', 'rating', 'timestamp', 'size', 'user_attr','model_attr', 'category', 'brand'
                 if cuitem["user_id"] == groupname:
                     
                     userdict['size'] = cuitem['size']
                     userdict['model_attr'] = cuitem['model_attr']
                     userdict['user_attr'] = cuitem['user_attr']
                     userdict['user_id'] = groupname
                     if cuitem['item_id'] == itemid:
                         userdict['brand'] = cuitem['brand']
                         userdict['category'] = cuitem['category']
                         userdict['movie_id'] = itemid
                         userdict['rating'] = cuitem['rating']
                     else:
                         for item in itemprofile:
                             if item["item_id"] == itemid:
                                 userdict['brand'] = item['brand']
                                 userdict['category'] = item['category']
                                 userdict['rating'] = 0
                                 userdict['movie_id'] = itemid
                 userdict["hist_movie_id"] = " ".join((str(id) for id in behavior1))
                 userdict['hist_len'] = len(behavior)
                 user_behavior.append(userdict)
                 continue

finalResult=pd.DataFrame(user_behavior)
finalResult.to_csv('modtrain.csv',sep='\t')
#
# all_data_dict=df_movies.to_dict(orient="records")
#
# user_behavior_df=pd.DataFrame(user_behavior)


print("preprocess done!!!")
```
